<h1>Random Sampling</h1>

```python
import numpy as np
import pandas as pd

data = pd.read_csv("rnaseq.csv")

first_col = data.iloc[:, :2]

rest = data.iloc[: , 1:]

#pick n random columns
samp = rest.sample(n=100,axis='columns')

df = pd.concat([first_col,samp], axis = 1)

df.reset_index(drop=True, inplace=True)

df.to_csv("samples100.csv", index = False)
```

```python
#clean by columns
#ignore this
import numpy as np
import pandas as pd

data = pd.read_csv("rnaseq.csv", skiprows = 1).T
data.drop_duplicates()

def remove_outliers(df,columns,n_std):
    for col in columns:
        #print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mean+(n_std*sd))]
        
    return df

new_data = data.iloc[1: , :]
no_outliers = remove_outliers(new_data,data.head() ,4)
print(no_outliers.shape)
```

```python
no_outliers.to_csv("no_outliers_3.csv")
```
